### SONAR (SOund Navigation And Ranging)

##### There is war going on between two countries and the enemy country has planted mines under the ocean. We now have to identify the mine and deactivate it. For this we are going to use SONAR and investigate using Machine Learning technique.

##### SONAR is a devise which uses ultrasonic waves to measure the distance, direction and speed of the object and help mariners to investigate about the object

```python
#import sonar data using pandas
import pandas as pd
data = pd.read_csv(r"E:\Suhani\Notes\Data Science\Data\sonar_csv.csv")
data
```

```python
#About the data
print("Sonar_data_information: ",data.info())
```

```python
#number of Mines and Rocks detected
print("Number of Mines and Rockes: ",data["Class"].value_counts())
```

```python
#mean of Each target (mine and rock)
data.groupby("Class").mean()
```

```python
#Classify the data into feature(as x) and target(as y) 
x=data.drop(['Class'],axis=1)
y=data['Class']
print(x)
print(y)
```

```python
#tranform class: Mine to 0 and Rock to 1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_y=le.fit_transform(y)
#print the tranformed 'Class' and its shape
print("Encoded y: ",encoded_y)
print("Shape of encoded y: ",encoded_y.shape)
```

```python
#Split the data(feature and target) to train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,encoded_y,test_size=0.1,stratify=y,random_state=1)
```

```python
#Use Logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
print("With respect to Logistic Regression")

#accuracy on training data
from sklearn.metrics import accuracy_score
y_train_prediction=model.predict(x_train)
training_data_accuracy = accuracy_score(y_train_prediction,y_train)
print("Accuracy of training data: ",training_data_accuracy)

#accuracy on testing data
y_test_prediction=model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test_prediction,y_test)
print("Accuracy of testing data: ",testing_data_accuracy)

#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn import metrics
cm = confusion_matrix(y_train,y_train_prediction)
print("Confusion matrix: ",cm)
```

```python
#Svm
from sklearn import svm
model1 = svm.SVC(kernel='linear')

#train the model
model1.fit(x_train, y_train)

#Accuracy Score of train data
from sklearn.metrics import accuracy_score
y_train_prediction1 = model1.predict(x_train)
train_accuracy_score1=accuracy_score(y_train_prediction1,y_train)
print(train_accuracy_score1)

#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn import metrics
cm1 = confusion_matrix(y_train,y_train_prediction1)
print("Confusion matrix: ",cm1)

#Accuracy Score of test data
y_test_prediction1 = model1.predict(x_test)
test_accuracy_score1=accuracy_score(y_test_prediction1,y_test)
print(test_accuracy_score1)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()

#train the model
model2.fit(x_train, y_train)

#Accuracy Score of train data
from sklearn.metrics import accuracy_score
y_train_prediction2 = model2.predict(x_train)
train_accuracy_score2=accuracy_score(y_train_prediction2,y_train)
print(train_accuracy_score2)

#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn import metrics
cm2 = confusion_matrix(y_train,y_train_prediction2)
print("Confusion matrix: ",cm1)

#Accuracy Score of test data
y_test_prediction2 = model2.predict(x_test)
test_accuracy_score2=accuracy_score(y_test_prediction2,y_test)
print(test_accuracy_score2)
```

```python
print("Based on Accuracy score it is clear that, it is better to use Logistic Regression.")
```
